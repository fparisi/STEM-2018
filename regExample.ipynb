```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
plt.style.use('seaborn-white')
import seaborn as sns; sns.set()
from scipy import stats

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
## This is just example data I made up
## Use your real data for the modeling
X_yourdata = np.random.random(100)
y_yourdata = np.random.random(100) + 0.5 * X_yourdata
```

```python
plt.scatter(X_yourdata, y_yourdata)
```

```python
## The scikit-learn funciton 'train_test_split' splits the data into a training set 
## and test set
## Build your model on the training data and test on your test data
Xtrain, Xtest, ytrain, ytest = train_test_split(X_yourdata, y_yourdata)
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(Xtrain[:, np.newaxis], ytrain)
```

```python
yfit = model.predict(Xtest[:, np.newaxis])
```

```python
plt.scatter(Xtrain, ytrain)
plt.plot(Xtest, yfit, "-r");
```

```python
resid = ytest-yfit
```

```python
plt.scatter(Xtest,resid)
```

```python
import statsmodels as sm
import statsmodels.regression.linear_model as lm

X = X_yourdata
Y = y_yourdata
# adds a 1 to the data so the model can estimate a coefficient term
X = sm.tools.tools.add_constant(X)  

# fit the model and display results 
model = lm.OLS(Y, X)
results = model.fit()
results.summary()
```
